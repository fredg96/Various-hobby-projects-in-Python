```python
import pandas as pd
import seaborn as sb
import numpy as np
from sklearn import ensemble, preprocessing, model_selection, metrics, decomposition, neighbors 
import os
from joblib import dump
pd.set_option('display.max_columns', 70)
pd.set_option('display.max_rows', 70)
```

```python
currentFolder = os.getcwd()
allData = pd.read_csv(currentFolder + '/data/resultingData/merged_data.csv')
allData
```

```python
data = allData.drop(columns = ['Unnamed: 0', 'Unnamed: 0_x', 'District', 'Member', 'Prior experience', 'Education', 
                               'Assumed office', 'Residence', 'Born', 'Spouse', 'Childrens', 'Years in office',
                               'Age when elected', 'Unnamed: 0_y', 'Race Total population','Birth Total Population', 'State'])
data
```

```python
features = data.columns.tolist()
for i in range(len(features)):
    print(type(data.loc[1,features[i]]))
```

```python
data['Party'] = data['Party'].replace({'Republican':0,'Democratic':1,'Libertarian':2})
target = data.pop('Party')
target
```

```python
data = pd.concat([data,pd.get_dummies(data['Most Employees'], prefix='employer')],axis=1)
data = pd.concat([data,pd.get_dummies(data['Largest Payroll'], prefix='paying')],axis=1)
data = pd.concat([data,pd.get_dummies(data['Most Establishments'], prefix='established')],axis=1)
data = data.drop(columns = ['Most Employees', 'Largest Payroll', 'Most Establishments'])
```

```python
data
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(data, target, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
```

Before normalizing the data we would remove any outliers however in this case the outliers aren't faulty data and hence we'll keep them since they do display some statisticall data.

```python
x_train.loc[:,'Total Population':'Total Occupation'] = x_train.loc[:,'Total Population':'Total Occupation'].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
x_test.loc[:,'Total Population':'Total Occupation'] = x_test.loc[:,'Total Population':'Total Occupation'].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
model = ensemble.RandomForestClassifier(max_depth = 12)#Create model
print("Fitting data to random forest.")
model.fit(x_train, y_train) #Train model
```

```python
scores = model_selection.cross_val_score(model, x_train, y_train, cv=10) #Do cross validation
print("\n Cross validation score: \n")
print(scores) #Report scores of cross validation
print("\n Mean cv score: ", np.mean(scores))
```

```python
prediction = model.predict(x_test)
accuracy = metrics.accuracy_score(y_test,prediction)
print(accuracy)
print("\n Confussion matrix:\n %s" % metrics.confusion_matrix(y_test, prediction))
```

So the results from just applying a random forest classifier results in an accuracy of 72.4% on the test set. The next thing we'll do is look at how to improve the performance. 

First we'll look at decreasing the dimensionality of the data by using principal component analysis which basically tries to project the data into a subspace of lower dimensions while keeping as much of the variance as possible.

```python
pca = decomposition.PCA(0.95) #n_components hardcodes the dimension, setting the argument to a = (0,1) makes the algorithm calculate 
                              # n components explaining a*100% of the variance.
pca.fit(x_train)
x_train_low_dim = pca.transform(x_train)
x_test_low_dim = pca.transform(x_test)
```

```python
print(pca.components_.shape)
print(pca.explained_variance_)
```

So we can reduce our data from the original 59 dimensions into 22 dimensions while still explaining 95% of the variance.

Now we'll train and test a new model using this lowere dimensional data.

```python
model = ensemble.RandomForestClassifier(max_depth = 12)#Create model
print("Fitting data to random forest.")
model.fit(x_train, y_train) #Train model

scores = model_selection.cross_val_score(model, x_train, y_train, cv=10) #Do cross validation
print("\n Cross validation score: \n")
print(scores) #Report scores of cross validation
print("\n Mean cv score: ", np.mean(scores))

prediction = model.predict(x_test)
accuracy = metrics.accuracy_score(y_test,prediction)
print(accuracy)
print("\n Confussion matrix:\n %s" % metrics.confusion_matrix(y_test, prediction))
```

Performing the PCA lowered the accuracy when using the same structure for the random forest classifier so now we'll try and find a better model by optimising the hyperparameters.

```python
n_estimators = [int(x) for x in np.linspace(start = 1, stop = 1000, num = 100)]
max_features = ['auto', 'sqrt', 'log2']
max_depth = [int(x) for x in np.linspace(1, 110, num = 11)]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

randomSearchGrid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
model = ensemble.RandomForestClassifier()
modelRandom = model_selection.RandomizedSearchCV(estimator = model, param_distributions = randomSearchGrid,
                               n_iter = 100, cv = 10, verbose=2, random_state=42, n_jobs = -1)
modelRandom.fit(x_train, y_train)
modelRandom.best_params_
```

```python
bestRandom = modelRandom.best_estimator_
prediction = bestRandom.predict(x_test)
accuracy = metrics.accuracy_score(y_test,prediction)
print(accuracy)
print("\n Confussion matrix:\n %s" % metrics.confusion_matrix(y_test, prediction))
```

So after the random search we didn't improved the accuracy. Next up is doing a local grid search around the rprevious found parameters

```python
searchGrid = {
    'bootstrap': [False],
    'max_depth': [40, 45, 50, 55, 60, 65],
    'max_features': ['sqrt'],
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [1, 2, 3, 4],
    'n_estimators': [290, 300, 313, 320, 330]
}
model = ensemble.RandomForestClassifier()
modelGrid = model_selection.GridSearchCV(estimator = model, param_grid = searchGrid, 
                          cv = 10, n_jobs = -1, verbose = 2)
modelGrid.fit(x_train_low_dim, y_train)
modelGrid.best_params_
```

```python
bestGrid = modelGrid.best_estimator_
prediction = bestGrid.predict(x_test_low_dim)
accuracy = metrics.accuracy_score(y_test,prediction)
print(accuracy)
print("\n Confussion matrix:\n %s" % metrics.confusion_matrix(y_test, prediction))
```

So here we saw a small improvement compared to the untuned parameters. Naturally we could continue this redifenment as much as we want.

```python
searchGrid = {
    'bootstrap': [False],
    'max_depth': [58, 59, 60, 61, 62],
    'max_features': ['sqrt'],
    'min_samples_leaf': [1, 2],
    'min_samples_split': [3, 4, 5, 6],
    'n_estimators': [ 310, 313, 315]
}

model = ensemble.RandomForestClassifier()
modelGrid = model_selection.GridSearchCV(estimator = model, param_grid = searchGrid, 
                          cv = 10, n_jobs = -1, verbose = 2)
modelGrid.fit(x_train_low_dim, y_train)
print(modelGrid.best_params_)

bestGrid = modelGrid.best_estimator_
prediction = bestGrid.predict(x_test_low_dim)
accuracy = metrics.accuracy_score(y_test,prediction)
print(accuracy)
print("\n Confussion matrix:\n %s" % metrics.confusion_matrix(y_test, prediction))
```

In this notebook we used a built in function to perform random search and then grid search but ofcourse we could do it systematically changing a parameter at a time and plot how the accuracy is affected, though it should be noted that we'll then tune the parameters to work well with each others with the hope that it willl produce an atleast local optimal classifier.

Finally we'll compare our tuned random forest classifier with a tuned k-nearest neighbour classifier.

```python
grid = {'n_neighbors': [int(x) for x in np.linspace(start = 1, stop = 20, num = 20)],
        'weights':['uniform', 'distance'],
        'algorithm': ['auto'],
        'p':[int(x) for x in np.linspace(start = 1, stop = 5, num = 5)]
        }
model = neighbors.KNeighborsClassifier()
modelGrid = model_selection.GridSearchCV(estimator = model, param_grid = grid, 
                          cv = 10, n_jobs = -1, verbose = 2)

modelGrid.fit(x_train_low_dim, y_train)
print(modelGrid.best_params_)

bestGrid = modelGrid.best_estimator_
prediction = bestGrid.predict(x_test_low_dim)
accuracy = metrics.accuracy_score(y_test,prediction)
print(accuracy)
print("\n Confussion matrix:\n %s" % metrics.confusion_matrix(y_test, prediction))
```
